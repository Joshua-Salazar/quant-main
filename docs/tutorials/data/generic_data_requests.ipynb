```python
import os
import sys
from pathlib import Path
module_path = Path('.').absolute().parents[4]
if str(module_path) not in sys.path:
    sys.path.append(str(module_path))

from capstone.quant.infrastructure.generic_data_container import GenericDataRequest, GenericDataContainer, GenericDataFrameFlatFileDataSource
```

```python
# GenericDataFrameFlatFileDataSource provides a way to load any dataframe like data from a flat csv file
# just as any other data type, this fits into the -- DataRequest, DataSource, DataContainer -- triparty structure

# first create a generic data request
data_request = GenericDataRequest()

# then create a GenericDataFrameFlatFileDataSource, with the source csv file's name specified in the constructor
file_name = "capstone/quant/test_data/source/index_returns.csv"
file_name = module_path.resolve().joinpath(file_name).absolute()
data_source = GenericDataFrameFlatFileDataSource(file_name=file_name)

# calling datasource's initialize function will return an GenericDataContainer object, which contains the retrieved data
data_container = data_source.initialize(data_request)

# calling the get_data() interface function will get back the actual dataframe
data = data_container.get_data()
```
